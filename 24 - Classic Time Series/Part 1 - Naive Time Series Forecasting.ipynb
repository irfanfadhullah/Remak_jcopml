```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data/daily_min_temp.csv',index_col='Date',parse_dates=['Date'])
df.head()
```

# Resampling Data Time Series

```python
df.Temp.plot(figsize=(15,3))
```

```python
df.resample('w').mean().plot() #w = weekly
```

```python
df.resample('m').mean().plot() #m = monthly
```

```python
df=df.resample('m').mean()
df
```

```python
df.Temp.plot(figsize=(15,5))
```

# Naive Forecasting

### One-shift

```python
pred = df.copy()
pred = pred.shift()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```

### Mean as Forecast

```python
pred=df.copy()
pred.Temp = [df.loc[:date].Temp.mean() for date in df.index]
pred = pred.shift()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```

### Moving Average /Rolling Mean Forecasting

```python
pred=df.copy()
pred.Temp = df.rolling(3).mean() #window nya rata2 3 hari terakhir
pred = pred.shift()
```

```python
plt.figure(figsize=(15,5))
plt.plot(df,'b-')
plt.plot(pred,'r-')
```
